```python
import pprint as pp
import requests 
import json
from datetime import date
import datetime
import pandas as pd
from time import mktime
from datetime import datetime, timedelta
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('shares.json') as g:
    file_data = json.load(g)
```

```python
sharesdf= pd.DataFrame(file_data)
sharesdf
```

```python
nsf= pd.DataFrame({"Newspaper": sharesdf['newspaper'], "Date":sharesdf['date'], "shares": sharesdf['shares']})
```

```python
l1 =[]
for i in range(nsf.shape[0]):
    
    try:
        l1.append(list(nsf.shares[i].values())[0])
    except:
        l1.append(0)
nsf['fb_sh']=l1
nsd = nsf[['Newspaper','Date','fb_sh']]
nsd
```

```python
ab = nsd.groupby(["Date","Newspaper"]).sum()
```

```python
ab= ab.reset_index()
ab = ab.sort_values(by = ["Newspaper","Date"])
ab.to_csv(r'share_news_jan_april.csv')
```

```python
sumdf = pd.DataFrame(ab.groupby("Newspaper"))
```
